````python
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

RANDOM_STATE = 42
rng = np.random.default_rng(seed=RANDOM_STATE)
````

# Model

````python
class Model(nn.Module):
    def __init__(self, layer_sizes, output_bound=None, dropout_rate=0.2):
        super(Model, self).__init__()
        if output_bound is not None:
            self.out_shift = output_bound[0]
            self.out_scale = output_bound[1] - output_bound[0]
        else:
            self.out_shift = 0
            self.out_scale = 1
        layers = []
        for in_dim, out_dim in zip(layer_sizes[:-2], layer_sizes[1:-1]):
            layers.append(nn.Linear(in_dim, out_dim))
            layers.append(nn.ReLU())
            layers.append(nn.Dropout(dropout_rate))
        layers.append(nn.Linear(*layer_sizes[-2:]))
        if output_bound is not None:
            layers.append(nn.Sigmoid())
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        out = self.network(x)
        return self.out_scale * out + self.out_shift
````

# Model training function

````python
def train_model(model, train_loader, val_loaders, criterion, optimizer, num_epochs, device, report_every=10):
    model.to(device)
    history = {'train_loss': []}
    for i in range(len(val_loaders)):
        history[f"val_loss_{i}"] = []
    s = time.time()

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        for batch_x, batch_y in train_loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)

            optimizer.zero_grad()
            outputs = model(batch_x)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()

            train_loss += loss.item() * batch_x.size(0)

        train_loss /= len(train_loader.dataset)
        history['train_loss'].append(train_loss)

        model.eval()
        for i, val_loader in enumerate(val_loaders):
            val_loss = 0.0
            with torch.no_grad():
                for batch_x, batch_y in val_loader:
                    batch_x, batch_y = batch_x.to(device), batch_y.to(device)
                    outputs = model(batch_x)
                    loss = criterion(outputs, batch_y)
                    val_loss += loss.item() * batch_x.size(0)
            history[f"val_loss_{i}"].append(val_loss / len(val_loader.dataset))

        if (epoch + 1) % report_every == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.2e}, dt: {time.time()-s}, " +
                  ", ".join(f"Val Loss {i}: {history[f'val_loss_{i}'][-1]:.2e}" for i in range(len(val_loaders))))
            s = time.time()

    return history
````

````python
def plot_training_history(history, log_loss=False):
    """Plot training and validation loss"""
    plt.figure(figsize=(10, 5))
    plt.plot(history['train_loss'], label='train loss')
    i = 0
    while True:
        try:
            plt.plot(history[f'val_loss_{i}'], label=f'Validation Loss {i}')
            i += 1
        except KeyError:
            break
    plt.xlabel('epoch')
    if log_loss:
        plt.yscale('log')
        plt.ylabel('log(loss)')
    else:
        plt.ylabel('loss')
    plt.legend()
    plt.show()
````

# Model train / eval

````python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
````

Notes to self on encoding for `train_val_test_split`:
- pretty jank but will work well enough for our purposes
- format is `AAABCDD`
    - `AAA` is (1-p) as percentage, e.g. `100` is the completely undiluted lattice
    - `B` is `is_lr` (just `0` or `1` for now)
    - `C` is lattice type
    - `DD` is the index of the disorder configuration; currently we only use `00-09` but I'm adding enough for up to 100 configs if desired
- nice property of this is that the 0% dilution lattices can be checked with `enc >= 1000000`
- simple arithmetic to test for other conditions, e.g.
    - `enc % 100` gives disorder config index

````python
def train_val_test_split(X, Y, config_indices, train_frac=0.7, val_frac=0.15, holdout_val_frac=0.5):
    """
    does a compliated train-val-test split because we want to test both generalization across walks on previously-seen
    disorder, and generalization to unseen disorder configurations. returns 10 (!) different datasets, namely:
    - X_train, y_train: training data, obviously
    - X_val_1, y_val_1: validation data, same disorder configs as training data
    - X_val_2, y_val_2: validation data, one completely held-out disorder config
    - X_test_1, y_test_1: test data, same disorder configs as training data
    - X_test_2, y_test_2: testing data, same held-out disorder config as val_2
    specification of split sizes is a bit complicated...
    - one disorder config for each nonzero-dilution lattice type held out, regardless of size
        - this set is split according to holdout_val_frac to make val_2 and test_2
    - among the remaining disorder configs, split according to train_frac and val_frac to produce the training
        as well as val_1 / test_1 datasets

    return order:
        X_train, (X_val_1, X_val_2), (X_test_1, X_test_2), [same order for y]
    """
    
    y_enc = Y[:,1]*10000 + Y[:,2]*1000 + Y[:,0]*100 + config_indices

    # we hold out one disorder configuration to check generalization on an unseen disorder realization
    is_config_1 = (y_enc % 100 == 1)
    
    reserve_indices = np.where(is_config_1)[0]
    nonreserve_indices = np.where(~is_config_1)[0]

    y_enc_reserved = y_enc[reserve_indices]
    y_enc_nonreserved = y_enc[nonreserve_indices]
    idx_val_2, idx_test_2, _, _ = train_test_split(reserve_indices, y_enc_reserved, stratify=y_enc_reserved, train_size=holdout_val_frac, random_state=RANDOM_STATE)
    idx_train, idx_testval, _, _ = train_test_split(nonreserve_indices, y_enc_nonreserved, stratify=y_enc_nonreserved, train_size=train_frac, random_state=RANDOM_STATE)

    y_enc_testval = y_enc[idx_testval]
    idx_test_1, idx_val_1, _, _ = train_test_split(idx_testval, y_enc_testval, stratify=y_enc_testval, test_size=(val_frac/(1-train_frac)), random_state=RANDOM_STATE)

    # verify non-overlapping indices
    check_indices = np.zeros(y_enc.size, dtype=int)
    for idx in (idx_train, idx_val_1, idx_val_2, idx_test_1, idx_test_2):
        check_indices[idx] += 1
    bad = np.where(check_indices != 1)[0]
    if len(bad) > 0:
        print("WARNING: overlapping or missing indices after split")
        print(f"# of missing indices: {len(np.where(check_indices == 0)[0])}")
        print(f"# of overcounted indices: {len(np.where(check_indices > 1)[0])}")
    
    return (
        X[idx_train],
        (X[idx_val_1], X[idx_val_2]),
        (X[idx_test_1], X[idx_test_2]),
        Y[idx_train],
        (Y[idx_val_1], Y[idx_val_2]),
        (Y[idx_test_1], Y[idx_test_2]),
    )
````

````python
data = np.load('featurized_data.npz')
X = data['features']
Y = data['labels']
config_indices = data['config_indices']

X = X.astype(np.float32)

N_FEATURES = X.shape[1]
````

# select which feature to train on:

too lazy to do this properly, so set the `SELECT` flag to decide what kind of model gets trained...

````python
# 0 = lattice type; 1 = p; 2 = is_lr
SELECT = 2

LABEL_NAMES = ['lattice_type', 'p', 'is_lr']
IS_REGRESSION = [False, True, False]
OUTPUT_BOUNDS = [None, (0,100), None]
NETWORK_PARAMS = [
    [N_FEATURES, 64, 4],
    [N_FEATURES, 64, 64, 64, 1],
    [N_FEATURES, 10, 2],
]
LABEL_DTYPE = [np.int64, np.float32, np.int64]
DROPOUTS = [0.1, 0.1, 0.0]
LRS = [1e-3, 1e-3, 1e-3]
WDS = [1e-5, 1e-5, 1e-5]
EPOCHS = [30, 75, 5]
CLASS_NAMES = [
    np.load('data_packed.npz')['lattice_names'],
    None,
    ["nearest-neighbor", "long-range"]
]

print(f"training to predict {LABEL_NAMES[SELECT]}")
Y = Y.astype(LABEL_DTYPE[SELECT])
````

````python
X_train, X_vals, X_tests, y_train, y_vals, y_tests = \
    train_val_test_split(X, Y, config_indices)

y_train = y_train[:,SELECT]
y_vals = [y_val[:,SELECT] for y_val in y_vals]
y_tests = [y_test[:,SELECT] for y_test in y_tests]

if IS_REGRESSION[SELECT]:
    y_train = y_train[:,None]
    y_vals = [y_val[:,None] for y_val in y_vals]
    y_tests = [y_test[:,None] for y_test in y_tests]
````

````python
train_dataset = TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train))
val_datasets = [TensorDataset(torch.from_numpy(X_val), torch.from_numpy(y_val)) for X_val, y_val in zip(X_vals, y_vals)]

batch_size = 128
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loaders = [DataLoader(vd, batch_size=batch_size, shuffle=False) for vd in val_datasets]

print(f'Number of training batches: {len(train_loader)}')
print(f'Number of validation batches: {list(map(len, val_loaders))}')
````

````python
model = Model(
    layer_sizes=NETWORK_PARAMS[SELECT],
    dropout_rate=DROPOUTS[SELECT],
    output_bound=OUTPUT_BOUNDS[SELECT],
)
print(model)
print(f'\nTotal parameters: {sum(p.numel() for p in model.parameters())}')
````

````python
criterion = nn.MSELoss() if IS_REGRESSION[SELECT] else nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LRS[SELECT], weight_decay=WDS[SELECT])
print(f"using criterion: {criterion}")
````

# Actual training loop

````python
num_epochs = EPOCHS[SELECT]
history = train_model(
    model=model,
    train_loader=train_loader,
    val_loaders=val_loaders,
    criterion=criterion,
    optimizer=optimizer,
    num_epochs=num_epochs,
    device=device,
    report_every=1
)
plot_training_history(history, log_loss=True)
````

* If classification: print overall error rate and confusion matrix
* If regression: print some summary statistics on the error distribution, as well as error histogram and CDF

````python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(15,6))
model.eval()

if not IS_REGRESSION[SELECT]:
    with torch.no_grad():
        for i, (X_test, y_test, ax) in enumerate(zip(X_tests, y_tests, (ax1, ax2))):
            logits = model(torch.from_numpy(X_test).to(device)).cpu().numpy()
            predictions = np.argmax(logits, axis=1)
            print(f"test set {i} error rate: {(predictions != y_test).sum() / len(y_test):.1e}")
            cm = confusion_matrix(y_test, predictions)
            n_points_tot = cm.sum()
            cm = cm / n_points_tot
            labels = CLASS_NAMES[SELECT]
            sns.heatmap(cm, annot=True, cmap='Blues', xticklabels=labels, yticklabels=labels, ax=ax)
            ax.set_ylabel('Actual')
            ax.set_xlabel('Predicted')
            ax.set_title(f"Test set {i} ({n_points_tot} points)")

else:   
    percentiles = np.linspace(85,100,501,endpoint=True)  
    COLORS = [
        mcolors.to_rgb(c) for c in [
            'tab:blue',
            'tab:red',
        ]
    ]
    errors_list = []
    with torch.no_grad():
        for i, (X_test, y_test, ctuple) in enumerate(zip(X_tests, y_tests, COLORS)):
            X_test_torch = torch.from_numpy(X_test).to(device)
            predictions = model(X_test_torch).cpu().numpy()
            errors = (predictions - y_test)
            errors_list.append(errors)
            print(f"SUMMARY STATISTICS, TEST SET {i}:")
            print(f"number of points: {len(errors)}")
            print(f"MSE: {(errors**2).mean():.2f}")
            print(f"MAE: {abs(errors).mean():.2f}")
            print(f"max. abs. err.: {abs(errors).max():.2f}")
            print("")
            ax1.hist(
                errors,
                label=f"set {i}",
                range=(-25,25),
                histtype='stepfilled',
                facecolor=ctuple+(0.5,),
                edgecolor=ctuple,
                density=True
            )
            ax1.set_xlabel("error")
            ax1.set_ylabel("density")
    
            ax1.set_title("error distributions")
            ax2.set_title("absolute error, percentiles")
            ax2.plot(percentiles, np.percentile(abs(errors), percentiles), label=f"set {i}", c=ctuple)
            ax2.set_xlabel("percentile")
            ax2.set_ylabel("absolute error")
            ax2.grid(axis='x')
            ax1.legend()
            ax2.legend()
        ERR_THRESH = 5
        for i, err in enumerate(errors_list):
            big_err = (abs(err)>ERR_THRESH).sum()
            print(f"set {i} |err| > {ERR_THRESH} count: {big_err:<3} ({100*big_err/len(err):.2f}%)")

plt.savefig(f"Plots/model_perf_{LABEL_NAMES[SELECT]}.png")
plt.show()
````

````python
model_name = f"model_{LABEL_NAMES[SELECT]}.pth"
torch.save(model.state_dict(), model_name)
print(f"Model saved to {model_name}")
````
